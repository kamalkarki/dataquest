```python
import pandas as pd 
data = pd.read_csv('/home/kamlesh/AnacondaProjects/Dataquest/Data/AmesHousing.txt',delimiter='\t')
data.head()
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
train = pd.DataFrame()
test = pd.DataFrame()

def train_and_test(data):
    train = data[:1460]
    test = data[1460:]
    
    numeric_train = train.select_dtypes(include=['integer','float']).columns
    numeric_test = test.select_dtypes(include=['integer', 'float']).columns

    train = train[numeric_train].dropna()
    test = test[numeric_test].dropna()
    print(train.columns)
    features = train.columns.drop('SalePrice')
    
    lr = linear_model.LinearRegression()
    lr.fit(train[features],train["SalePrice"])
    predictions = lr.predict(test[features])
    
    mse = mean_squared_error(test["SalePrice"], predictions)
    rmse = mse ** 0.5
    return rmse

def transform_features(data):
    return data

def select_features():
    features = ['Gr Liv Area','SalePrice']
    return train[features]
```

```python
transform_df = transform_features(data)
rmse = train_and_test(transform_df)

print(rmse)

```
