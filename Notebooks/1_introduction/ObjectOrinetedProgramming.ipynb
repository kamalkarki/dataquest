```python
class Trial(object):
    def __init__(self, datarow):
        self.efficiency = float(datarow[0])
        self.individual = int(datarow[1])
        self.chopstick_length = int(datarow[2])

class Chopstick(object):
    def __init__(self, length):
        self.length = length
        self.trials = []
        for index,row in chopsticks.iterrows():
            if int(row[2]) == self.length:
                self.trials.append(Trial(row)) 
    def elements(self):
        return len(self.trials)
    def avg_efficiency(self ):
        sum_eff = 0
        for row in self.trials:
            sum_eff += row.efficiency
        return sum_eff / self.elements()   
    
as210 = Chopstick( 210).avg_efficiency() 
#print(len(as210.trials))
print(as210)
```
